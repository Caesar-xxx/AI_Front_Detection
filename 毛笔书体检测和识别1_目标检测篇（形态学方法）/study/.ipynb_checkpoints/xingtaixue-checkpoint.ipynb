```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
img = cv2.imread('./test_imgs/j.png')
```

```python
img.shape
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
gray.shape
```

```python
plt.imshow(gray,cmap='gray')
```

```python
#侵蚀
#作用：去除白色噪点，将两个连起来的形状打散
```

```python
kernel = np.ones((3,3),dtype=np.int8)
erosion = cv2.erode(gray.copy(),kernel,iterations=1)
```

```python
erosion.shape
```

```python
plt.imshow(erosion,cmap='gray')
```

```python
kernel = np.ones((5,5),dtype=np.int8)
erosion1 = cv2.erode(gray.copy(),kernel,iterations=1)
```

```python
plt.imshow(erosion1,cmap='gray')
```

```python
kernel = np.ones((5,5),dtype=np.int8)
erosion3 = cv2.erode(gray.copy(),kernel,iterations=2)
```

```python
plt.imshow(erosion3,cmap='gray')
```

```python
fig,(ax1,ax2,ax3,ax4) = plt.subplots(1,4,figsize=(20,8),sharex=True,sharey=True)
ax1.axis('off')
ax1.imshow(gray.copy(),cmap='gray')
ax1.set_title('original img')

ax2.axis('off')
ax2.imshow(erosion,cmap='gray')
ax2.set_title('3*3 img')

ax3.axis('off')
ax3.imshow(erosion1,cmap='gray')
ax3.set_title('5*5 img')

ax4.axis('off')
ax4.imshow(erosion3,cmap='gray')
ax4.set_title('5*5 2 img')
```

```python
#膨胀
#扩大白色边缘，跟在侵蚀操作后去噪点，把两个分开的部分连接起来
```

```python
plt.imshow(gray.copy(),cmap='gray')
```

```python
# 3*3,1
kernel = np.ones((3,3),dtype=np.int8)
dilation0 = cv2.dilate(gray.copy(),kernel,iterations=1)

# 5*5,1
kernel = np.ones((5,5),dtype=np.int8)
dilation1 = cv2.dilate(gray.copy(),kernel,iterations=1)

# 5*5,2
kernel = np.ones((5,5),dtype=np.int8)
dilation2 = cv2.dilate(gray.copy(),kernel,iterations=2)

fig,(ax1,ax2,ax3,ax4) = plt.subplots(1,4,figsize=(20,8),sharex=True,sharey=True)
ax1.axis('off')
ax1.imshow(gray.copy(),cmap='gray')
ax1.set_title('original img')

ax2.axis('off')
ax2.imshow(dilation0,cmap='gray')
ax2.set_title('3*3 img')

ax3.axis('off')
ax3.imshow(dilation1,cmap='gray')
ax3.set_title('5*5 img')

ax4.axis('off')
ax4.imshow(dilation2,cmap='gray')
ax4.set_title('5*5 2 img')

```

```python
#opening,张开
#侵蚀+膨胀
#主要是清除噪点
img = cv2.imread('./test_imgs/cv.png')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
gray.shape
```

```python
original = plt.imshow(gray,cmap='gray')
```

```python
kernel = np.ones((10,10),dtype=np.int8)
original1 = cv2.morphologyEx(gray.copy(),cv2.MORPH_OPEN,kernel)
plt.imshow(original1,cmap='gray')
```

```python
kernel = np.ones((12,12),dtype=np.int8)
original2 = cv2.morphologyEx(gray.copy(),cv2.MORPH_OPEN,kernel)
plt.imshow(original2,cmap='gray')
```

```python
kernel = np.ones((15,15),dtype=np.int8)
original3 = cv2.morphologyEx(gray.copy(),cv2.MORPH_OPEN,kernel)
plt.imshow(original3,cmap='gray')
```

```python
fig,(ax1,ax2,ax3,ax4) = plt.subplots(1,4,figsize=(20,8),sharex=True,sharey=True)
ax1.axis('off')
ax1.imshow(gray.copy(),cmap='gray')
ax1.set_title('original img')

ax2.axis('off')
ax2.imshow(original1,cmap='gray')
ax2.set_title('original1 img')

ax3.axis('off')
ax3.imshow(original2,cmap='gray')
ax3.set_title('original2 img')

ax4.axis('off')
ax4.imshow(original3,cmap='gray')
ax4.set_title('original3 img')
```

```python
#闭合：closing
#先膨胀在侵蚀，主要用于闭合主体内的小洞，或者一些黑色的点
```

```python
kernel = np.ones((10,10),dtype=np.int8)
original2 = cv2.morphologyEx(gray.copy(),cv2.MORPH_CLOSE,kernel)
plt.imshow(original2,cmap='gray')
```
