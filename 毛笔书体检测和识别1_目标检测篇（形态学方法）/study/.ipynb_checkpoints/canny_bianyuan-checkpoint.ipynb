```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
img = cv2.imread('./test_imgs/pumpkin.jpg')
```

```python
img_fixed = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_fixed.shape
```

```python
plt.imshow(img_fixed)
```

```python
edges1 = cv2.Canny(img.copy(),100,200)
edges2 = cv2.Canny(img.copy(),50,200)
edges3 = cv2.Canny(img.copy(),50,100)
```

```python
fig,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(20,8),sharex=True,sharey=True)
ax1.axis('off')
ax1.imshow(edges1,cmap='gray')
ax1.set_title('edges1 img')

ax2.axis('off')
ax2.imshow(edges2,cmap='gray')
ax2.set_title('edges2 img')

ax3.axis('off')
ax3.imshow(edges3,cmap='gray')
ax3.set_title('edges3 img')

```
